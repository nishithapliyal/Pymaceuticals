### Analysis

Capomuline and Ramicane have lower final tumor volumes than Infubinol and Ceftamin. Taking into account that the tumor volume started off as the same size, this suggests that Capomuline and Ramicane are more effective in reducing tumor volumes.

The correlation coefficient between mouse weight and the average tumor volume is 0.84, which is relatively high, suggesting that mouse weight and average tumor volume are relatively correlated. Which means that mice with a higher weight have larger tumor volumes than mice with smaller weights.

Out of all the drugs, Capomuline and Ramicane had the lowest standard deviations for tumor volumes, at 4.99 and 4.85 respectively. This suggests that the tumor volumes associated with the aforementioned drugs are within a tighter range around the mean tumor volumes per drug.

```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
from scipy.stats import linregress 

# Study data files
mouse_metadata_path = "data/Mouse_metadata.csv"
study_results_path = "data/Study_results.csv"

# Read the mouse data and the study results
mouse_metadata = pd.read_csv(mouse_metadata_path)
study_results = pd.read_csv(study_results_path)

# Combine the data into a single DataFrame
combined_df = pd.merge(study_results, mouse_metadata, how = 'outer', on = 'Mouse ID')

# Display the data table for preview
combined_df.head(5)
```

```python
# Checking the number of mice.
num_mice = combined_df['Mouse ID'].nunique()
num_mice
```

```python
# Our data should be uniquely identified by Mouse ID and Timepoint
# Get the duplicate mice by ID number that shows up for Mouse ID and Timepoint. 
duplicatemice = combined_df[combined_df.duplicated(subset=['Mouse ID', 'Timepoint'])]['Mouse ID'].unique()
duplicatemice
```

```python
# Optional: Get all the data for the duplicate mouse ID. 
duplicatedata = combined_df[combined_df['Mouse ID'] == duplicatemice[0]]
duplicatedata
```

```python
# Create a clean DataFrame by dropping the duplicate mouse by its ID.
cleaned_df = combined_df[combined_df['Mouse ID'].isin(duplicatemice)==False]
cleaned_df.head()
```

```python
# Checking the number of mice in the clean DataFrame.
num_mice_cleaned = cleaned_df['Mouse ID'].nunique()
num_mice_cleaned
```

## Summary Statistics

```python
# Generate a summary statistics table of mean, median, variance, standard deviation, and SEM of the tumor volume for each regimen

# Use groupby and summary statistical methods to calculate the following properties of each drug regimen: 
# mean, median, variance, standard deviation, and SEM of the tumor volume. 
# Assemble the resulting series into a single summary DataFrame.

mean = cleaned_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].mean()
median = cleaned_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].median()
var = cleaned_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].var()
stdev = cleaned_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].std()
sem = cleaned_df.groupby('Drug Regimen')['Tumor Volume (mm3)'].sem()

tumor_dict = {'Mean Tumor Volume': mean, 'Median Tumor Volume': median, 'Tumor Volume Variance': var,
              'Tumor Volume Std. Dev.': stdev, 'Tumor Volume Std. Err.': sem}
summary_stats = pd.DataFrame(tumor_dict)
summary_stats
```

```python
# A more advanced method to generate a summary statistics table of mean, median, variance, standard deviation,
# and SEM of the tumor volume for each regimen (only one method is required in the solution)

# Using the aggregation method, produce the same summary statistics in a single line
aggreg_summary = cleaned_df.groupby('Drug Regimen')[['Tumor Volume (mm3)']].agg([
    'mean', 'median', 'var', 'std', 'sem'])
aggreg_summary
```

## Quartiles, Outliers and Boxplots

```python
# Calculate the final tumor volume of each mouse across four of the treatment regimens:  
# Capomulin, Ramicane, Infubinol, and Ceftamin

# Start by getting the last (greatest) timepoint for each mouse
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

treatment_df = cleaned_df[cleaned_df['Drug Regimen'].isin(treatments)]
max_timepts = treatment_df.groupby('Mouse ID')['Timepoint'].max()

# # Merge this group df with the original DataFrame to get the tumor volume at the last timepoint
merged = pd.merge(treatment_df, max_timepts, on = ['Mouse ID', 'Timepoint'], how = 'right').reset_index(drop=True)
merged.head()
```

```python
# Put treatments into a list for for loop (and later for plot labels)
treatments = ['Capomulin', 'Ramicane', 'Infubinol', 'Ceftamin']

# Create empty list to fill with tumor vol data (for plotting)
tumor_vol_data = []

# Calculate the IQR and quantitatively determine if there are any potential outliers. 
for drug in treatments:
    
    # Locate the rows which contain mice on each drug and get the tumor volumes
    drug_df = merged[merged['Drug Regimen'] == drug]
    tumor_vol_data.append(drug_df['Tumor Volume (mm3)']) 
    
    # add subset 
    # Determine outliers using upper and lower bounds
    quartiles = drug_df['Tumor Volume (mm3)'].quantile([0.25,0.5,0.75])
    lower = quartiles[0.25]
    upper = quartiles[0.75]
    iqr = upper - lower
    lower_bound = lower - 1.5*iqr
    upper_bound = upper + 1.5*iqr

    outliers = []
    for x in drug_df['Tumor Volume (mm3)']:
        if (x < lower_bound) | (x > upper_bound):
            outliers.append(x)

    print(f"The IQR for {drug} is {iqr.round(2)}")
    print(f"{drug}'s potential outliers: {outliers}")
```

```python
# Generate a box plot that shows the distrubution of the tumor volume for each treatment group.
cap_tumor_vol = tumor_vol_data[0]
ram_tumor_vol = tumor_vol_data[1]
inf_tumor_vol = tumor_vol_data[2]
ceft_tumor_vol = tumor_vol_data[3]

fig, ax = plt.subplots()
data = [cap_tumor_vol, ram_tumor_vol, inf_tumor_vol, ceft_tumor_vol]

ax.boxplot(data, tick_labels=treatments, flierprops=dict(marker='o', markerfacecolor='r', markersize=13, linestyle='none'))
ax.set_ylabel('Final Tumor Volume (mm3)')
plt.show()
```

## Line and Scatter Plots

```python
# Generate a scatter plot of mouse weight vs. the average observed tumor volume for the entire Capomulin regimen
capomulin_whole = cleaned_df[cleaned_df['Drug Regimen'] == 'Capomulin']
capomulin_avg = capomulin_whole.groupby('Mouse ID')[['Weight (g)', 'Tumor Volume (mm3)']].mean()
capomulin_avg.plot('Weight (g)', 'Tumor Volume (mm3)', kind = 'scatter', ylabel = 'Average Tumor Volume (mm3)')
plt.show()
```

## Correlation and Regression

```python
# Calculate the correlation coefficient and a linear regression model 
# for mouse weight and average observed tumor volume for the entire Capomulin regimen

corr_coef, p_value = st.pearsonr(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])
corr_coef = corr_coef.round(2)
print(f'The correlation between mouse weight and the average tumor volume is {corr_coef}')

slope, intercept, rvalue, pvalue, stderr = linregress(capomulin_avg['Weight (g)'], capomulin_avg['Tumor Volume (mm3)'])
reg_vals = capomulin_avg['Weight (g)'] * slope + intercept

capomulin_avg.plot('Weight (g)', 'Tumor Volume (mm3)', kind = 'scatter', ylabel = 'Average Tumor Volume (mm3)')
plt.plot(capomulin_avg['Weight (g)'], reg_vals, color = 'r')
plt.show()
```
